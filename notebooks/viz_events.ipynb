```python
import json
import folium
json_file = '/Users/veamit/dev/deutsche-telekom/data/events_of_interest.json'


with open(json_file, 'r') as f:
    events = json.load(f)
```

```python
events[1]['lat']
```

```python
# Create a map centered around the average location of the events
latitudes = [event['lat'] for event in events if 'lat' in event]
longitudes = [event['lng'] for event in events if 'lng' in event]
```

```python
map_center = [sum(latitudes) / len(latitudes), sum(longitudes) / len(longitudes)]
my_map = folium.Map(location=map_center, zoom_start=5)
```

```python
# Add markers for each event
for event in events:
    location = event['location']
    popup_text = f"""
        <b>{event['name']}</b><br>
        Date: {event['date']}<br>
        Type: {event['event_type']}
    """
    folium.Marker(
        location=[event['lat'], event['lng']],
        popup=folium.Popup(popup_text, max_width=250),
        tooltip=event['name']
    ).add_to(my_map)
```
